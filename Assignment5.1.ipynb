```python
import pandas as pd
url='/content/netflix_titles.csv'
data=pd.read_csv(url)
summary=data.describe()
print(summary)
```

```python
import matplotlib.pyplot as plt
plt.hist(data['release_year'].dropna(),bins=20,edgecolor='black',color='skyblue')
plt.title('release_year Distribution')
plt.xlabel('relese_year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['type'].value_counts().plot(kind='bar',edgecolor='black', color='gray')
plt.title('Type Distribution')
plt.xlabel('movie type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['release_year'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of release_year')
plt.ylabel('relese_year')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data[['release_year']].dropna())
plt.show()
```

```python
correlation_matrix = data[['release_year']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='type', data=data, palette='Set2')
plt.title('type Count')
plt.xlabel('type')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='type', y='release_year', data=data, palette='muted')
plt.title('Release Year Distribution by Type')
plt.show()
```
